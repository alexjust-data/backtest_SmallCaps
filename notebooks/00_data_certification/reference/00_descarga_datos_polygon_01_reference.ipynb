  - te basta con tickers_2005_2026_upper_clean.parquet

  Si quieres remapear parte de esos 200 con criterio:

  - entonces sí hace falta este artefacto de revisión

  - el remapped_review solo sirve si quieres rescatar algunos tickers por validación manual contra

## Auditoria de Dividends contra el universo

Este bloque valida especificamente `D:\reference\dividends` contra el universo objetivo actual.

Objetivos:

- medir cobertura real por ticker del universo
- inspeccionar el schema real de dividends
- separar tickers con dividend data de tickers sin dividend data
- medir cantidad de eventos de dividendos y rango temporal
- dejar una muestra operativa para revisar casos concretos


````python
from pathlib import Path
import pandas as pd
from IPython.display import display

UNIVERSE_CSV = Path(r"C:\TSIS_Data\v1\backtest_SmallCaps\data\reference\universe_pti\tickers_2005_2026_upper_excluding_ohlcv_1m_missing_vs_daily_ge_2B_or_null.csv")
DIVIDENDS_ROOT = Path(r"D:\reference\dividends")

u = pd.read_csv(UNIVERSE_CSV)
u['ticker'] = u['ticker'].astype(str).str.upper().str.strip()
u = u[~u['ticker'].isin(['', 'NAN', 'NONE'])].copy()
u_unique = u[['ticker']].drop_duplicates().reset_index(drop=True)

METADATA_ONLY_COLS = {'ticker', '_dataset', '_ingested_utc'}
DATE_CANDIDATES = ['ex_dividend_date', 'exDate', 'ex_date', 'exDividendDate']

ticker_dirs = sorted([p for p in DIVIDENDS_ROOT.glob('ticker=*') if p.is_dir()])
records = []
sample_schema_any = None
sample_schema_real = None

for d in ticker_dirs:
    ticker = d.name.split('ticker=', 1)[-1].upper().strip()
    files = sorted(d.glob('*.parquet'))
    if not files:
        records.append({
            'ticker': ticker,
            'has_dividend_file': False,
            'has_dividend_data': False,
            'rows': 0,
            'real_data_rows': 0,
            'first_ex_date': pd.NaT,
            'last_ex_date': pd.NaT,
            'columns': None,
            'schema_kind': 'no_file',
        })
        continue

    fp = files[0]
    df = pd.read_parquet(fp)
    cols = list(df.columns.astype(str))
    if sample_schema_any is None:
        sample_schema_any = cols

    data_cols = [c for c in cols if c not in METADATA_ONLY_COLS]
    has_real_data = len(data_cols) > 0
    schema_kind = 'real_dividend_data' if has_real_data else 'metadata_only_placeholder'

    if has_real_data and sample_schema_real is None:
        sample_schema_real = cols

    ex_col = next((c for c in DATE_CANDIDATES if c in df.columns), None)
    first_ex = pd.NaT
    last_ex = pd.NaT
    if ex_col is not None:
        ex = pd.to_datetime(df[ex_col], errors='coerce')
        if ex.notna().any():
            first_ex = ex.min()
            last_ex = ex.max()

    records.append({
        'ticker': ticker,
        'has_dividend_file': True,
        'has_dividend_data': has_real_data,
        'rows': int(len(df)),
        'real_data_rows': int(len(df)) if has_real_data else 0,
        'first_ex_date': first_ex,
        'last_ex_date': last_ex,
        'columns': ','.join(cols),
        'schema_kind': schema_kind,
    })

div_cov = pd.DataFrame(records)
audit = u_unique.merge(div_cov, on='ticker', how='left')
audit['has_dividend_file'] = audit['has_dividend_file'].fillna(False).astype(bool)
audit['has_dividend_data'] = audit['has_dividend_data'].fillna(False).astype(bool)
audit['rows'] = pd.to_numeric(audit['rows'], errors='coerce').fillna(0).astype(int)
audit['real_data_rows'] = pd.to_numeric(audit['real_data_rows'], errors='coerce').fillna(0).astype(int)
audit['schema_kind'] = audit['schema_kind'].fillna('missing_in_dividends_root')

summary = pd.DataFrame([{
    'universe_unique_tickers': int(len(u_unique)),
    'ticker_dirs_in_dividends_root': int(len(ticker_dirs)),
    'tickers_with_dividend_file': int(audit['has_dividend_file'].sum()),
    'tickers_with_real_dividend_data': int(audit['has_dividend_data'].sum()),
    'tickers_with_metadata_only_placeholder': int((audit['has_dividend_file'] & ~audit['has_dividend_data']).sum()),
    'tickers_missing_in_dividends_root': int((~audit['has_dividend_file']).sum()),
    'pct_with_real_dividend_data': round(100 * audit['has_dividend_data'].mean(), 2),
    'total_real_dividend_rows_in_universe': int(audit['real_data_rows'].sum()),
}])

print('=== DIVIDENDS COVERAGE VS UNIVERSE ===')
display(summary)

schema_mix = (audit.groupby('schema_kind', dropna=False).agg(tickers=('ticker', 'nunique')).reset_index().sort_values('tickers', ascending=False))
print('=== SCHEMA MIX ===')
display(schema_mix)

print('=== SAMPLE SCHEMA (ANY) ===')
print(sample_schema_any if sample_schema_any is not None else 'No schema found')
print('=== SAMPLE SCHEMA (REAL DIVIDEND DATA) ===')
print(sample_schema_real if sample_schema_real is not None else 'No real dividend-data schema found')

with_data = audit[audit['has_dividend_data']].copy().sort_values(['real_data_rows', 'ticker'], ascending=[False, True])
without_real = audit[~audit['has_dividend_data']].copy().sort_values('ticker')

print('=== TOP TICKERS WITH REAL DIVIDEND DATA ===')
display(with_data[['ticker', 'real_data_rows', 'first_ex_date', 'last_ex_date']].head(50))

print('=== TICKERS WITHOUT REAL DIVIDEND DATA ===')
display(without_real[['ticker', 'has_dividend_file', 'schema_kind']].head(50))

print('=== SAMPLE OF REAL DIVIDEND FILES ===')
for t in with_data['ticker'].head(5).tolist():
    fp = DIVIDENDS_ROOT / f'ticker={t}'
    files = sorted(fp.glob('*.parquet'))
    if not files:
        continue
    df = pd.read_parquet(files[0])
    print(f'--- {t} ---')
    display(df.head(10))
````
