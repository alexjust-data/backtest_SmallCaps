build_tickers_2005_2026_upper_remapped_review.py

Artefactos creados:

- tickers_2005_2026_upper_remapped_review.parquet
- tickers_2005_2026_upper_remapped_review.csv
- tickers_2005_2026_upper_remapped_review.summary.json

Resultado real:

- rows_total: 200

Qué incluye:

- ticker_original
- ticker_suggested
- candidate_exists_in_upper
- full_name
- request_date
- msg
- datos del ticker sugerido en upper
- datos del ticker sugerido en full
- manual_review_status = pending
- review_note

---


Ahora mismo tienes 200 tickers que:

  - están en tu universo operativo
  - fallan en overview
  - muchos parecen W.I. o símbolos transitorios
  - pero tú sospechas que algunos deberían remapearse al ticker “base” sin sufijo

  Ese código sirve para esto:

  - separar los 200 problemáticos
  - proponer un ticker base candidato
  - dejar una tabla de revisión manual antes de tocar el universo real

  Lo que evita:

  - hacer remapeos automáticos peligrosos
  - por ejemplo convertir un preferente, when-issued o unidad en un common stock que no corresponde

  En corto:

  - upper_clean sirve para excluir lo que ahora rompe
  - remapped_review sirve para revisar, uno a uno, cuáles de esos excluidos sí quieres convertir a ticker
    base

  Si no quieres revisión manual y prefieres expulsarlos todos:

  - entonces este código no hace falta para la operativa inmediata
  - te basta con tickers_2005_2026_upper_clean.parquet

  Si quieres remapear parte de esos 200 con criterio:

  - entonces sí hace falta este artefacto de revisión

  Mi lectura práctica:

  - para seguir ya, usa upper_clean
  - el remapped_review solo sirve si quieres rescatar algunos tickers por validación manual contra
    TradeStation/u otra fuente

````python
from pathlib import Path
import json
import pandas as pd

SCRIPT = Path(r"C:\TSIS_Data\v1\backtest_SmallCaps\notebooks\cell_code\build_tickers_2005_2026_upper_remapped_review.py")

exec(SCRIPT.read_text(encoding="utf-8"), globals(), globals())

summary = json.loads(Path(r"C:\TSIS_Data\v1\backtest_SmallCaps\data\reference\universe_pti\tickers_2005_2026_upper_remapped_review.summary.json").read_text(encoding="utf-8"))
review = pd.read_parquet(r"C:\TSIS_Data\v1\backtest_SmallCaps\data\reference\universe_pti\tickers_2005_2026_upper_remapped_review.parquet")

print(json.dumps(summary, indent=2, ensure_ascii=False))
display(review.head(30))
display(review[review["candidate_exists_in_upper"] == True].head(30))
````
